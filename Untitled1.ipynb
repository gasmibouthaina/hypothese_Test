```python
import pandas as pd
df = pd.read_csv('RateMyProfessor_Sample data.csv')
```

```python
#Visualisation de La note en fonction du genre
sns.boxplot(x='gender', y='asian', data=df)
plt.xlabel('gender')
plt.ylabel('asian')
plt.title("evaltuation de l'enseignement en fonction du sexe  de l'enseignement ")
plt.show()
```

```python
# Histogramme de La variable asian
sns.displot(df['asian'], kde=False);
```

```python
# Test de normalité de Jarque-Bera
stats.jarque_bera(df['asian'])
```

```python
# Test de normalité de Shapiro-wilk
stats.shapiro(df['asian'])
```

```python
# Test d'égalité des variances
stats.levene(df[df.gender =='female']['asian'], df[df.gender =='male']['asian'] )
```

```python
# Test de student
stats.test_ind(df[df.gender =='female']['asian'], df[df.gender =='male']['asian'], equal_var=True)
```
